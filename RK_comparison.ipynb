```python
import numpy as np
import matplotlib.pyplot as plt

tinc = 0.1
alpha = 0.1
def calculate_mse(tend):
    
    tlist = np.arange(0.0, tend, tinc)

    # Compute solutions using different methods
    rk4_solution = RK4(tinc, tend)[1]
    ark4_solution = ARK4(tinc, tend, alpha=alpha)[1]
    radau_solution = radau(tinc, tend)[1]


    # Calculate Mean Square Error (MSE) between Radau and other methods
    mse_radau_rk4 = np.mean((radau_solution - rk4_solution) ** 2)
    mse_radau_ark4 = np.mean((radau_solution - ark4_solution) ** 2)


    return mse_radau_rk4, mse_radau_ark4

# Initialize lists to store MSE values and tend values
mse_radau_rk4 = []
mse_radau_ark4 = []

tend_values = list(range(1, 8000, 400))

# Calculate MSE for different tend values
for index, tend in enumerate(tend_values):
    print(index/(len(tend_values)))
    mse_rk4, mse_ark4 = calculate_mse(tend)
    mse_radau_rk4.append(mse_rk4)
    mse_radau_ark4.append(mse_ark4)
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(tend_values, mse_radau_rk4, label='RK4')
plt.plot(tend_values, mse_radau_ark4, label='Adaptive RK4', linestyle='dashed')


plt.title(f'Mean Square Error (MSE) of phi_dot vs. Simulation Time  | tinc={tinc}  | alpha={alpha} | Radau as comparison ground')
plt.xlabel('Simulation Time (tend)')
plt.ylabel('Mean Square Error (MSE)')
plt.legend()
plt.grid(True)
plt.show()

print("MSE between Radau and RK4: ", mse_radau_rk4[-1])
print("MSE between Radau and ARK4:", mse_radau_ark4[-1])
```

# How does alpha choice alter the MSE for phi_dot (tinc = konst.)

```python
import numpy as np
import matplotlib.pyplot as plt


def calculate_mse(tend, alpha):
    tinc = 0.01
    tlist = np.arange(0.0, tend, tinc)

    # Compute solutions using different methods
    rk4_solution = RK4(tinc, tend)[1]
    ark4_solution = ARK4(tinc, tend, alpha)[1]
    radau_solution = radau(tinc, tend)[1]

    # Calculate Mean Square Error (MSE) between Radau and other methods
    mse_radau_rk4 = np.mean((radau_solution - rk4_solution) ** 2)
    mse_radau_ark4 = np.mean((radau_solution - ark4_solution) ** 2)

    return mse_radau_rk4, mse_radau_ark4

# Initialize lists to store MSE values and tend values
alpha_values = np.arange(0.01, 0.2, 0.01)
mse_radau_rk4 = np.zeros(len(alpha_values))
mse_radau_ark4 = np.zeros(len(alpha_values))

tend = 16000

# Calculate MSE for different alpha values
for i, alpha in enumerate(alpha_values):
    print(i/(len(alpha_values)))
    mse_rk4, mse_ark4 = calculate_mse(tend, alpha)
    mse_radau_rk4[i] = mse_rk4
    mse_radau_ark4[i] = mse_ark4


# print("MSE between Radau and RK4: ", mse_radau_rk4)
# print("MSE between Radau and ARK4:", mse_radau_ark4)
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(alpha_values, mse_radau_rk4, label='RK4', linestyle='dashed')
plt.plot(alpha_values, mse_radau_ark4, label='ARK4')

plt.title(f'Mean Square Error (MSE) of phi_dot vs. Alpha   tend={tend}, tinc=0.01')
plt.xlabel('Alpha')
plt.ylabel('Mean Square Error (MSE) of phi_dot')
plt.legend()
plt.grid(True)
plt.show()
```
